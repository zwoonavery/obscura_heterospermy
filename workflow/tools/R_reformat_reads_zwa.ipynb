### Set up data

```r
## load in data and reformat, finding a mean for each species
daff_reads <- read_delim("../../workflow/abundances/Daff/stringtie_gene_counts.txt", skip = 1) %>%
    mutate(Gene_ID = str_replace(Geneid, "gene-", "")) %>%
    rename_with(~ str_remove(., "HISAT-2/Daff/"), everything()) %>%
    rename_with(~ str_remove(., "_Daff.csorted.hisat2.bam"), everything()) %>%
    select(Gene_ID, starts_with("Daff")) %>%
    rowwise() %>%
    mutate("Drosophila_affinis" = mean(c_across(starts_with("Daff")))) %>%
    select(c(Gene_ID, "Drosophila_affinis"))
dazt_reads <- read_delim("../../workflow/abundances/Dazt/stringtie_gene_counts.txt", skip = 1) %>%
    mutate(Gene_ID = str_replace(Geneid, "gene-", "")) %>%
    rename_with(~ str_remove(., "HISAT-2/Dazt/"), everything()) %>%
    rename_with(~ str_remove(., "_Dazt.csorted.hisat2.bam"), everything()) %>%
    select(Gene_ID, starts_with("Dazt")) %>%
    rowwise() %>%
    mutate("Drosophila_azteca" = mean(c_across(starts_with("Dazt")))) %>%
    select(c(Gene_ID, "Drosophila_azteca"))
dbif_reads <- read_delim("../../workflow/abundances/Dbif/stringtie_gene_counts.txt", skip = 1) %>%
    mutate(Gene_ID = str_replace(Geneid, "gene-", "")) %>%
    rename_with(~ str_remove(., "HISAT-2/Dbif/"), everything()) %>%
    rename_with(~ str_remove(., "_Dbif.csorted.hisat2.bam"), everything()) %>%
    select(Gene_ID, starts_with("Dbif")) %>%
    rowwise() %>%
    mutate("Drosophila_bifasciata" = mean(c_across(starts_with("Dbif")))) %>%
    select(c(Gene_ID, "Drosophila_bifasciata"))
dgua_reads <- read_delim("../../workflow/abundances/Dgua/stringtie_gene_counts.txt", skip = 1) %>%
    mutate(Gene_ID = str_replace(Geneid, "gene-", "")) %>%
    rename_with(~ str_remove(., "HISAT-2/Dgua/"), everything()) %>%
    rename_with(~ str_remove(., "_Dgua.csorted.hisat2.bam"), everything()) %>%
    select(Gene_ID, starts_with("Dgua")) %>%
    rowwise() %>%
    mutate("Drosophila_guanche" = mean(c_across(starts_with("Dgua")))) %>%
    select(c(Gene_ID, "Drosophila_guanche"))
dmad_reads <- read_delim("../../workflow/abundances/Dmad/stringtie_gene_counts.txt", skip = 1) %>%
    mutate(Gene_ID = str_replace(Geneid, "gene-", "")) %>%
    rename_with(~ str_remove(., "HISAT-2/Dmad/"), everything()) %>%
    rename_with(~ str_remove(., "_Dmad.csorted.hisat2.bam"), everything()) %>%
    select(Gene_ID, starts_with("Dmad")) %>%
    rowwise() %>%
    mutate("Drosophila_madeirensis" = mean(c_across(starts_with("Dmad")))) %>%
    select(c(Gene_ID, "Drosophila_madeirensis"))
dobs_reads <- read_delim("../../workflow/abundances/Dobs/stringtie_gene_counts.txt", skip = 1) %>%
    mutate(Gene_ID = str_replace(Geneid, "gene-", "")) %>%
    rename_with(~ str_remove(., "HISAT-2/Dobs/"), everything()) %>%
    rename_with(~ str_remove(., "_Dobs.csorted.hisat2.bam"), everything()) %>%
    select(Gene_ID, starts_with("Dobs")) %>%
    rowwise() %>%
    mutate("Drosophila_obscura" = mean(c_across(starts_with("Dobs")))) %>%
    select(c(Gene_ID, "Drosophila_obscura"))
dpse_reads <- read_delim("../../workflow/abundances/Dpse/stringtie_gene_counts.txt", skip = 1) %>%
    mutate(Gene_ID = str_replace(Geneid, "gene-", "")) %>%
    rename_with(~ str_remove(., "HISAT-2/Dpse/"), everything()) %>%
    rename_with(~ str_remove(., "_Dpse.csorted.hisat2.bam"), everything()) %>%
    select(Gene_ID, starts_with("Dpse")) %>%
    rowwise() %>%
    mutate("Drosophila_pseudoobscura" = mean(c_across(starts_with("Dpse")))) %>%
    select(c(Gene_ID, "Drosophila_pseudoobscura"))
dsub_reads <- read_delim("../../workflow/abundances/Dsub/stringtie_gene_counts.txt", skip = 1) %>%
    mutate(Gene_ID = str_replace(Geneid, "gene-", "")) %>%
    rename_with(~ str_remove(., "HISAT-2/Dsub/"), everything()) %>%
    rename_with(~ str_remove(., "_Dsub.csorted.hisat2.bam"), everything()) %>%
    select(Gene_ID, starts_with("Dsub")) %>%
    rowwise() %>%
    mutate("Drosophila_subobscura" = mean(c_across(starts_with("Dsub")))) %>%
    select(c(Gene_ID, "Drosophila_subobscura"))
```

```r
## combine the reads into one tsv
reads <- daff_reads %>%
         full_join(dazt_reads, by = "Gene_ID") %>%
         full_join(dbif_reads, by = "Gene_ID") %>%
         full_join(dgua_reads, by = "Gene_ID") %>%
         full_join(dmad_reads, by = "Gene_ID") %>%
         full_join(dobs_reads, by = "Gene_ID") %>%
         full_join(dpse_reads, by = "Gene_ID") %>%
         full_join(dsub_reads, by = "Gene_ID") %>%
         mutate(across(.cols = starts_with("D"), ~ case_when(is.na(.x) ~ 0, TRUE ~ .x))) %>%
         rownames_to_column("DESC") %>%
         mutate(DESC = paste0("transcript", DESC))

write_tsv(reads, "obscura_group_reads.tsv")
```
